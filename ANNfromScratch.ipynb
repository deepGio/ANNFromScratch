### Reminder:

An Artificial Neural Network (ANN) is a type of machine learning system that can generally learn complex functions to map an input data to an output data. There's a lot more, but we focus on this capability that is under the so called "supervised learning" area. We tell the network that "y" corresponds to the input "x" and give it a lot of examples of (x, y) pairs: the training data. Hopefully the network learns the way to correctly compute "y" starting from "x" so that in the future we can give it only x values and get back y values (predictions).

A typical ANN architecture is made of:

- **Input Layer** (the number of neurons are the number of features we want to include)
- One or more **Hidden Layers** (each composed by an arbitrary number of neurons)
- **Output Layer** (the final layer: neurons cardinality is the number of output features)

#### How ANN computes the y values from x values?

There's nothing more than matrices multiplication and activations: the **Forward Propagation**. Remind the image on top.

##### First hidden layer:

\begin{cases} z^{[1]}_{1} = w^{[1]}_{1}*x + b^{[1]} \\ 
z^{[1]}_{2} = w^{[1]}_{2}*x + b^{[1]} \\ 
z^{[1]}_{3} = w^{[1]}_{3}*x + b^{[1]} \\
z^{[1]}_{4} = w^{[1]}_{4}*x + b^{[1]}
\end{cases}

*the $.^{[i]}$ index stands for the i-th layer in the network  
*the $._{j}$ index stands for the j-th neuron of the considered layer

##### Applying activation:

The values returned by the previous equations are then passed through an activation function like sigmoid or tanh or ReLU.

\begin{cases} a^{[1]}_{1} = g(z^{[1]}_{1}) \\ 
a^{[1]}_{2} = g(z^{[1]}_{2}) \\ 
a^{[1]}_{3} = g(z^{[1]}_{3}) \\
a^{[1]}_{4} = g(z^{[1]}_{4})
\end{cases}

![image](images/activations.png)

The same is done for the last layer where instead of 4 neurons there is only one unit.  
It comes out that the only way to affect the prediction (y value) is tuning the inner parameters of the ANN, the so called W matrices (weights) and b vectors (biases).  
This tuning is automatically made by the **BackPropagation** algorithm by computing the gradients of a special function (the **cost function**) that represents the error between predicted Y and real Y and that considers also W and b terms.  
By getting the gradients with respect to all this parameters, we can update them in the direction of "minor cost".  
The magic is that BackPropagation algorithm is automatically managed by Tensorflow so we only need to build ANN Tensors and Operations for the Forward Propagation.

#### Let's start!


### Checking for dependencies and Loading Libraries

```python
try:
    import tensorflow as tf
except:
    print("You don't have tensorflow installed yet")
try:
    import numpy as np
except:
    print("You don't have numpy installed yet")
try:
    import matplotlib.pyplot as plt
except:
    print("You don't have matplotlib installed yet")
from IPython.display import display, clear_output, Markdown
clear_output()
```

### Let's generate some data: 
We use the following equation to create examples of (x, y) pairs:
<br>
## $$y=ax^3 + bx^2 + cx$$
<br>
The aim is to teach the network the function that maps x to the correct y. 

#### Tune your function parameters:

```python
a = 2
b = -3
c = -6
```

#### Our data:

```python
display(Markdown("### Function: $$y=("+str(a)+")x^3+("+str(b)+")x^2+("+str(c)+")x$$"))
x = (np.random.rand(1, 200) -0.5)*2
y_raw = (a*x)**3 + (b*x)**2 + c*x
y = y_raw/np.max(y_raw)
fig, ax = plt.subplots(ncols=2, figsize=[10,4], dpi=200)
ax[0].scatter(x, y_raw, s=2, label="Original Data")
ax[1].scatter(x, y, s=2, label="Normalized Data")
ax[0].set_xlabel("X")
ax[1].set_xlabel("XNorm")
ax[0].set_ylabel("Y")
ax[1].set_ylabel("YNorm")
ax[0].set_title("Polynomial function")
ax[1].set_title("Normalized polynomial function")
ax[0].grid()
ax[1].grid()
plt.show()
```

### Setting up network hyperparameters:

- **Batch Size**: the number of examples (x-y pairs) to be given to ANN simultaneously for one iteration.
- **Epochs**: the number of times the entire training set passes throught the network
- **Hidden Units**: number of neurons of the hidden layer

```python
batch_size = 2
epochs = 20
hidden_units = 12
```

### Preprocessing data: random shuffle and splitting into train/test sets

```python
# randomly splitting train/dev
indices = np.arange(0,x.shape[1])
np.random.shuffle(indices)
x_train = x[:,indices[:int(float(len(indices))*0.7)]]
x_test = x[:,indices[int(float(len(indices))*0.7):]]
y_train = y[:,indices[:int(float(len(indices))*0.7)]]
y_test = y[:,indices[int(float(len(indices))*0.7):]]
train_num_batches = int(float(x_train.shape[1])/batch_size)
test_num_batches = int(float(x_test.shape[1])/batch_size)
print("X Train shape: ", x_train.shape)
print("Y Train shape: ", y_train.shape)
print("X Test shape: ", x_test.shape)
print("Y Test shape: ", y_test.shape)
```

### Creating computational Graph 

```python
# Graph:

## Input Tensors: one for X values and one fot Y values: 

xx = tf.placeholder(tf.float32, shape=[1, batch_size], name='x')
yy = tf.placeholder(tf.float32, shape=[1, batch_size], name='y')

## HIDDEN LAYER:
# Initializing Weights (W) and biases (b) for the hidden layer:

W1 = tf.Variable(tf.random_normal(dtype=tf.float32, shape=[hidden_units,1]), name='W1')
b1 = tf.Variable(tf.zeros((hidden_units,1)), tf.float32, name='b1')

# Compute the matrices multiplication and apply RELU (Rectified Linear Unit) activation:

z1 = tf.add(tf.matmul(W1, xx), b1)
a1 = tf.nn.relu(z1)

## OUTPUT LAYER:
# Initializing Weights (W2) and biases (b2):

W2 = tf.Variable(tf.random_normal(dtype=tf.float32, shape=[1,hidden_units]), name='W2')
b2 = tf.Variable(tf.zeros((1,1)), tf.float32, name='b2')

# Compute matrices multiplication and apply TANH (hyperbolic tan) activation to the output layer:
z2 = tf.add(tf.matmul(W2, a1), b2)
y_hat = tf.tanh(z2)

# Compute loss function (difference between predicted Y (Y_HAT) and real Y):
cost = tf.reduce_mean(tf.square(yy - y_hat))

# Setting up Gradient Descend Optimizer to backpropagate the error and update weights and biases to minimize loss:

optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(cost, var_list=[W1,b1,W2,b2])

# This will initialize all variables and build the graph:
init = tf.global_variables_initializer()
```

### Your Graph representation on TensorBoard:

Tensorboard is a cool interface built on top of Tensorflow, to view computational graph schema, monitoring variables and losses during training time and many other things.

![image](images/tboard.png)

### Training ANN with Forward and BackPropagation

```python
slider = 0
slider_max = train_num_batches*epochs
losses = []
with tf.Session() as sess:
    sess.run(init)
    for m in range(0,epochs):
        loss = []
        for k in range(0,train_num_batches):
            x_batch = x_train[:, k*batch_size:k*batch_size+batch_size]
            y_batch = y_train[:, k*batch_size:k*batch_size+batch_size]
            sess.run(train, feed_dict={xx: x_batch, yy: y_batch})
            loss_ = sess.run(cost, feed_dict={xx: x_batch, yy: y_batch})
            loss.append(loss_)
            slider += 1
            clear_output(wait=True)
            display(Markdown("Iteration number: "+str(slider)+" of "+str(slider_max)))
        losses.append(np.mean(loss))
            
    plt.figure(figsize=[10,4], dpi=200)
    plt.plot(losses, linewidth=2)
    plt.title("Loss Function over training. Training Final Loss: "+str(losses[-1]))
    plt.grid()
    plt.xlabel("Number of Epochs")
    plt.show()
    plt.figure(figsize=[10,4], dpi=150)
    losses_test = []
    for k in range(test_num_batches):
        x_batch = x_test[:, k*batch_size:k*batch_size+batch_size]
        y_batch = y_test[:, k*batch_size:k*batch_size+batch_size]
        plt.scatter(x_batch[0], sess.run(y_hat, feed_dict={xx: x_batch})[0], c='blue', alpha=.5)
        plt.scatter(x_batch[0], y_batch[0], c='orange', alpha=.7)
        losses_test.append(sess.run(cost, feed_dict={xx: x_batch, yy: y_batch}))
    loss_test = np.mean(losses_test)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Test Mean Squared Error: "+str(loss_test))
    plt.grid()
    plt.legend(["PREDICTED", "REAL"])
    plt.show()
```
